### model representation

learn to implement the model $f_{w,b}$ linear regerssion with one variable

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train=np.array([1.0,2.0]) #1000 of sqft
y_train=np.array([300.0,500.0]) #1000s of dollars
```

```python
#m=len(x_train)  # we can use this way also to find the length of the traing data 
m=x_train.shape[0] # to find the total number of training
m
```

```python
# plt the graph
plt.scatter(x_train,y_train,marker='x',c='red')# plot data point
plt.title("house prices")
plt.ylabel("price(1000$)")
plt.xlabel("size(1000sqft)")
plt.show()
```

```python
def linear_regerssion_model(x,w,b):
    m=x.shape[0]
    f_wb=np.zeros(m)
    for i in range(m):
        f_wb[i]=(w*x[i])+b
    return f_wb 
```

```python
w=200
b=100
```

```python
t_f_wb=linear_regerssion_model(x_train,w,b)
```

```python
plt.plot(x_train,t_f_wb,c='blue', label='prediction')
plt.scatter(x_train,y_train,marker='x',c="red" ,label="actual values")
plt.ylabel("price in 1000s")
plt.xlabel("size in 1000sqft")
plt.title("house predicted price")
plt.legend()
plt.show()
```

```python
# now we have a model so we can use this to make predication

w=200
b=100
x_i=1.2  #1000sqft
cost=(w*x_i)+b

print(f"the price for the 1200sqft is ${cost:.0f}")
```
