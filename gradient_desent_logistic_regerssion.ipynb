### logistic gradient desent

```python
def gradient_logistic(X,y,w,b):
    m,n=X.shape
    dj_dw=np.zeros((n,)) #create a 1D array with n elements
    dj_db=0.
    for i in range(m):
        f_wb_i=sigmoid(np.dot(X[i],w)+b)
        err_i=f_wb_i-y[i]
        for j in range(n):
            dj_dw[j]=dj_dw[j]+(err_i*X[i,j])
        dj_db=dj_db+err_i

    dj_db=dj_db/m
    dj_dw=dj_dw/m

    return dj_dw,dj_db
```

```python
X_tmp = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y_tmp = np.array([0, 0, 0, 1, 1, 1])
w_tmp = np.array([2.,3.])
b_tmp = 1.
dj_dw_tmp,dj_db_tmp = gradient_logistic(X_tmp, y_tmp, w_tmp, b_tmp)
print(f"dj_db: {dj_db_tmp}" )
print(f"dj_dw: {dj_dw_tmp.tolist()}" )
```

```python
def gradient_desent_logistic(X,y,w_init,b_init,alpha,numb_itter):
    J_history=[] 
    w=copy.deepcopy(w_init)
    b=b_init
    for i in range(numb_itter):
        dj_dw,dj_db=gradient_logistic(X,y,w,b)
        w=w-alpha*dj_dw
        b=b-alpha*dj_db

        if i<100000:   # prevent resource exhaustion 
            J_history.append( compute_cost_logistic(X, y, w, b) )

        # Print cost every at intervals 10 times or as many iterations if < 10
        if i% math.ceil(numb_itter / 10) == 0:
            print(f"Iteration {i:4d}: Cost {J_history[-1]}   ")
    return w,b,J_history
        
```

```python
m,n=X_train.shape
np.zeros_like(n)
```

```python
X_train = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y_train = np.array([0, 0, 0, 1, 1, 1])
w_tmp  = np.zeros_like(X_train[0])
b_tmp  = 0.
alph = 0.1
iters = 10000

w_out, b_out, _ = gradient_desent_logistic(X_train, y_train, w_tmp, b_tmp, alph, iters) 
print(f"\nupdated parameters: w:{w_out}, b:{b_out}")
```

```python
# lets see the gradient desent graph

fig,ax = plt.subplots(1,1,figsize=(5,4))
# plot the probability 
plt_prob(ax, w_out, b_out)

# Plot the original data
ax.set_ylabel(r'$x_1$')
ax.set_xlabel(r'$x_0$')   
ax.axis([0, 4, 0, 3.5])
plot_data(X_train,y_train,ax)

# Plot the decision boundary
x0 = -b_out/w_out[0]
x1 = -b_out/w_out[1]
ax.plot([0,x0],[x1,0], c=dlc["dlblue"], lw=1)
plt.show()
```
