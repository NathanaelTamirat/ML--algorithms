```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
```

```python
w=tf.Variable(3.0)
x=1.0
y=1.0
alpha=0.01
oprimizer=tf.keras.optimizers.Adam(learning_rate=1e-1)
iteration=30
w_values = []
cost_values = []
for iter in range(iteration):
    with tf.GradientTape() as tape :
        fwb=w*x
        costJ=(fwb -y)**2
    
    [djdw]=tape.gradient(costJ,[w])

    w.assign_add(-alpha*djdw)
    w_values.append(w.numpy())
    cost_values.append(costJ.numpy())
```

```python
plt.scatter(w_values,cost_values)
plt.xlabel("w_values")
plt.ylabel("cost_values")
plt.show()
```
